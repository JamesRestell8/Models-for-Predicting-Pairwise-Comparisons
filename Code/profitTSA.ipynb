```python
def getListofResiduals(mu, var, n):
    return np.random.normal(mu, math.sqrt(var), n).tolist() 
```

```python
listOfProfits = []
for i in range(1000):
    thisProfits = []
    thisResiduals = getListofResiduals(0, 4.846, 2000)
    for j in range(2000):
        # simulate ARIMA(0,2,1)
        if j == 0:
            nextVal = thisResiduals[0]
        elif j == 1:
            nextVal = 2*thisProfits[j-1] + thisResiduals[j] - 0.9994536*thisResiduals[j-1]
        else:
            nextVal = 2*thisProfits[j-1] - thisProfits[j-2] + thisResiduals[j] - 0.9994536*thisResiduals[j-1]
        thisProfits.append(nextVal)
    listOfProfits.append(thisProfits)
                
```

```python
fig, ax = plt.subplots()
for profit in listOfProfits:
    ax.plot(profit)
```
